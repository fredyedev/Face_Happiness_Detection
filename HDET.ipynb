```python
from fer import FER
import cv2
```

```python
cap = cv2.VideoCapture('video.mp4')
length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'MP4V') #codec
out = cv2.VideoWriter('out.mp4', fourcc, 20.0, (width,  height))
counter=0
while cap.isOpened():

  ret, frame = cap.read()
  if not ret:
    print("Fineshed ...")
    break

  detector = FER(mtcnn=True)
  detsfer = detector.detect_emotions(frame)

  for detfer in detsfer:
    x,y,w,h = detfer['box']
    frame = cv2.rectangle(frame, (int(x), int(y)), (int(x+w), int(y+h)),
        (25,75,88), 3)
    if detfer['emotions']['happy'] >= 0.38 :
      frame = cv2.rectangle(frame, (int(x), int(y-30)),
        (int(x)+80,int(y)), (25,75,88), -1)
      frame = cv2.putText(frame, "Happy", (int(x), int(y-10)), 0, 0.75,
        (255, 255, 255), 2)
    else :
      frame = cv2.rectangle(frame, (int(x), int(y-30)),
        (int(x)+90,int(y)), (25,75,88), -1)
      frame = cv2.putText(frame, "Not happy", (int(x), int(y-10)), 0, 0.75,
        (255, 255, 255), 2)

  counter+=1
  print("Progress: ",counter,"/",length)
  out.write(frame)

cap.release()
out.release()
cv2.destroyAllWindows()
```
